## Setup

```python
import matplotlib.pyplot as plt
import random
```

## Make a Histogram

```python
numbers = [2, 2, 7, 5, 2, 6, 9, 1, 8]  # lots of numbers, all on one line 
print(numbers)
```

### Plot the histogram

```python
# make the histogram 
plt.hist(numbers, bins=3, color='blue', alpha=0.8)

# save the chart directly
plt.savefig('hist1.png')

# display chart
plt.show()
```

### Changed the number of bins, changed the color  

```python
plt.hist(numbers, bins=7, color='orange', alpha=0.8)

# display chart
plt.show()
```

### Sets the number of bins to one third of the number of items in each list

```python
def bins_one_third_list_size(x_list):
    """Returns the number of containers equal to one-third of the number of items in a list."""
    return round(len(x_list) / 3) 
    
```

```python
bins = bins_one_third_list_size(numbers)
print(f'Numbers of bins: {bins}')

# make the histogram 
plt.hist(numbers, bins=bins, color='lightblue', alpha=0.8)

# display chart
plt.show()
```
