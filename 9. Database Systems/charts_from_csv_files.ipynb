### Introduction

The data involved relate to some characteristics and habits of some students and are related to a single day of collection. The students wrote their answers on a spread sheet provided by a teacher. 

Questions: 
1. How many hours of sleep did you get last night? (as a float)
2. What is your height in metres? (as a float)
3. According to Google Maps, what is the distance in kilometres from your house to school? (as a float)

The data are synthetic, there was no real statistical survey.

### Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Load and process data

```python
datatable = np.genfromtxt(
    './Data/day3.csv',  # file name
    skip_header=1,  # skip the first line
    delimiter=',',  # this is a Comma Separated Value file 
    dtype='float',  # data type is float
    filling_values=0,  # in case a data element is missing
    usecols=(0, 1, 2))

print(datatable)
print(f'This table has {len(datatable)} rows')
```

```python
# Use subsetting techniques for create three separate arrays 
sleephours = datatable[:, 0]
heights = datatable[:, 1]
distances = datatable[:, 2]
 
print(f'Values of hours of sleep of students: \n{sleephours}\n')
print(f'Values of heights of students: \n{heights}\n')
print(f'Values of distances of students: \n{distances}\n')
```

### Plot Histogram of distances 

```python
# make the histogram 
plt.hist(distances, bins=10, rwidth=0.8, color='blue', alpha=0.6)

# make x-label and a title
plt.xlabel('Distance in kilometres')
plt.title('Student home-school distance histogram')

# save the chart directly
plt.savefig('./SavedCharts/histogram.png')

# display chart
plt.show()
```

### Plot scattergram of hours of sleep and height of students

```python
# make the scatter plot
plt.scatter(sleephours, heights, c='darkblue', alpha=0.6)

# make x-label, y-label and a title
plt.xlabel('Hours of sleep (h)')
plt.ylabel('Height in meters')
plt.title('Scatter plot between hours of sleep and height of students')

# save the chart directly
plt.savefig('./SavedCharts/scattergram.png')

# display chart
plt.show()
```
