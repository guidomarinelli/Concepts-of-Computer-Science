**Description:** Create a pie chart and slice the data.<br>

```python
import matplotlib.pyplot as plt
```

```python
# set up your lists
numlist = [8, 6, 5, 3]
namelist = ['freshmen', 'sophomores', 'juniors', 'seniors']
colorlist = ['red', 'green', 'pink', 'yellow']
explodelist = [0.1, 0.00, 0.0, 0.0]
```

### Plot the pie chart

```python
# make the pie chart
plt.pie(numlist, labels=namelist, autopct='%.2f%%', colors=colorlist,
        explode=explodelist, startangle=90)
plt.axis('equal')

# save the chart directly
plt.savefig('piechart.png')

# display chart
plt.show()
```

### We make changes to the chart, such as: 
* custom the match the name with numbers and colours 
* Offset of multiple slices using _explode_
* change the starting angle

```python
explodelist = [0.1, 0.01, 0.07, 0.01]
colorlist = ['lightblue', 'orange', 'green', 'red']

# make the pie chart
plt.pie(numlist, labels=namelist, autopct='%.2f%%', colors=colorlist,
        explode=explodelist, startangle=60)
plt.axis('equal')

# display chart
plt.show()
```

* Change the values to a decimal place

```python
# make the pie chart
plt.pie(numlist, labels=namelist, autopct='%.1f%%', colors=colorlist,
       explode=explodelist, startangle=60)
plt.axis('equal')

# display chart
plt.show()
```

### Making subsets of the namelist

```python
# Divide namelist into two subsets
subset1 = namelist[:2]
subset2 = namelist[2:]

print(f'First subset from namelist: {subset1}')
print(f'Second subset from namelist: {subset2}')
```
